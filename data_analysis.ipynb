```python
#importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
df = pd.read_csv('california_housing_dataset.csv')
df
```

### Analyzing the data

```python
df.info()
```

```python
df.describe()
```

### Removing the outliers

```python
from scipy import stats
df_clean = df.copy()
for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_clean[col] = df[col].clip(lower, upper)
df_clean.describe()
```

## Dealing with missing values

```python
df.isnull().sum()
```

```python
#Using median imputation
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median') 
df_imputed = pd.DataFrame(
    imputer.fit_transform(df_clean),
    columns=df_clean.columns
)
df = df_imputed
df
```

### We will remove the abnormal spikes

```python
print(df.skew())
print(df.corr()['target'].sort_values())
```

```python

for i in ["Population","AveBedrms","AveOccup"]:
    df[i]= np.log1p(df[i])

```

```python
#Checking if there is a need of applyiing log on "MedInc"
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].scatter(df['MedInc'], df['target'], alpha=0.1)
axes[0].set_title('Original MedInc vs Target')

axes[1].scatter(np.log1p(df['MedInc']), df['target'], alpha=0.1)
axes[1].set_title('Log MedInc vs Target')
```

```python

df = df[(df['target'] < 4.8)]  # catches the upper spike
df = df[np.abs(df['target'] -1.795)> 0.05 ]
df = df.reset_index(drop=True)

print(len(df))
sns.histplot(df['target'])
```
