### Loading the inbuilt sklearn dataset

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd 
data = fetch_california_housing()
df = pd.DataFrame(data.data, columns = data.feature_names)
df['target']= data.target
df

```

## Adding missing values


```python
import numpy as np

def add_missing_values(df, frac=0.1):
    df_copy = df.copy()
    for col in df_copy.columns:
        df_copy.loc[df_copy.sample(frac=frac).index, col] = np.nan
    return df_copy

df = add_missing_values(df, frac=0.1)

```

```python
df.to_csv('california_housing_dataset.csv',index= False)
```
